```python
#komolyabb pdftoimage 
from pdf2image import convert_from_path


# ✅ Provide absolute path to Poppler's bin folder
poppler_path = r"C:/Program Files/poppler-24.08.0/Library/bin"

# ✅ Convert PDF to PNG images
pdf_path = r"C:\Users\ebali\Downloads\01_Vez_Alk01_téglatest1 (1).pdf"  # Change to your actual PDF file
images = convert_from_path(pdf_path, dpi=200, poppler_path=poppler_path)

# ✅ Save images and store filenames
image_paths = []
for i, image in enumerate(images):
    image_path = f"page_{i+1}.png"
    image.save(image_path, "PNG")
    image_paths.append(image_path)  # Store filenames in a list

print("PDF successfully converted to PNG!")
print("Image paths:", image_paths)
import os

absolute_image_paths = [os.path.abspath(image_path) for image_path in image_paths]
print("Image paths:", absolute_image_paths)
```
